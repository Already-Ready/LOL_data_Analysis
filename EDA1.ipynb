* 각 랭크별 league 데이터 가져오기

```python
master = 'https://kr.api.riotgames.com/lol/league/v4/masterleagues/by-queue/RANKED_SOLO_5x5?api_key=' + api_key
```

```python
grandmaster = 'https://kr.api.riotgames.com/lol/league/v4/grandmasterleagues/by-queue/RANKED_SOLO_5x5?api_key=' + api_key
```

```python
challenger = 'https://kr.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5?api_key=' + api_key
```

```python
from pandas.io.json import json_normalize
```

```python
'''
각 랭크별로 데이터 요청
'''
r_master = requests.get(master)
r_grandmaster = requests.get(grandmaster)
r_challenger = requests.get(challenger)
```

```python
'''
각 랭크별로 df 생성
'''
master_league_df = pd.DataFrame(r_master.json())
grandmaster_league_df = pd.DataFrame(r_master.json())
challenger_league_df = pd.DataFrame(r_master.json())
```

```python
len(master_league_df) + len(grandmaster_league_df) + len(challenger_league_df)
```

```python
'''
랭크 df 통합
'''
concat_league_df = pd.concat([master_league_df, grandmaster_league_df, challenger_league_df],ignore_index=True)
concat_league_df
```

```python
'''
통합 df의 entries 컬럼을 풀어낸 entries df 생성
'''
concat_entries_df = pd.DataFrame(dict(concat_league_df['entries'])).T
concat_entries_df
```

```python
'''
entried df와 league df를 axis=1로 열끼리 결합해서 하나의 df생성
'''
concat_df = pd.concat([concat_league_df, concat_entries_df], axis=1)
concat_df = concat_df.drop(['entries'], axis=1) # entries를 풀어서 합친 결과물이므로 중복되는 칼럼 삭제
concat_df
```

* 유저의 고유한 accountId 를 위의 결과를 통해 수집

```python
concat_df_copy = concat_df.copy()
```

```python
for i in tnrange(len(concat_df_copy), desc='1st Loop'):
    try:
        user = 'https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/' + concat_df_copy['summonerName'].iloc[i] + '?api_key=' + api_key 
        r = requests.get(user)
        
        while r.status_code == 429:
            time.sleep(5)
            user = 'https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/' + concat_df_copy['summonerName'].iloc[i] + '?api_key=' + api_key 
            r = requests.get(user)
            
        account_id = r.json()['accountId']
        concat_df_copy.iloc[i, -1] = account_id
    
    except:
        pass
```

```python
# 마지막 칼럼에 accoundId를 넣어줬으므로 해당 칼럼 이름 변경
concat_df_copy.rename(columns={'wins':'accountId'}, inplace=True)
```

```python
# 추후 이용을 위해 저장
concat_df_copy.to_csv('data/saving_data/concat_data.csv', index=False, encoding='utf-8')
```
