### EDA4에서 만든 데이터는 boolean값, str값, int값이 사용되고있다.
### boolean, str값을 int로 매핑시켜주고 필요치 않은 칼럼은 drop하는 과정을 거친다.

```python
import numpy as np
import pandas as pd
import requests
import json
from tqdm import tnrange
import time
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_pickle('data/saving_data/df_info_EDA4')
```

```python
data.head()
```

* boolean 값을 0과 1로 바꿔주기 위해 boolean인 칼럼을 리스트로 만들자.

```python
bool_col_list = ['bluefirstBlood','bluefirstTower','bluefirstInhibitor','bluefirstBaron','bluefirstDragon',
                 'bluefirstRiftHerald',
                 'redfirstBlood','redfirstTower','redfirstInhibitor','redfirstBaron','redfirstDragon',
                 'redfirstRiftHerald',
                 '1stats.win','2stats.win','3stats.win','4stats.win','5stats.win',
                 '6stats.win','7stats.win','8stats.win','9stats.win','10stats.win']
```

* 다음과 같이 간단하게 True는 1로 False는 0으로 바꿔줄 수 있다.
* 다만, 아래의 방법은 NaN값이 존재하지 않아야 가능하므로 주의

```python
for i in bool_col_list:
    data[i] = data[i].astype('int')
```

```python
data.head()
```

```python
data.columns
```

* 소환사의 협곡에서 사용하지 않는 칼럼들은 삭제해준다.

```python
data.drop(['bluevilemawKills','bluedominionVictoryScore',
           'redvilemawKills','reddominionVictoryScore'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.bluewin = data.bluewin.replace({'Win': 1, 'Fail': 0})
data.redwin = data.redwin.replace({'Win': 1, 'Fail': 0})
```

```python
data.head()
```

* NaN 값들도 0으로 채워서 데이터 타입을 맞출 수 있도록 만들자.
* lane 정보의 NONE 값은 타입을 확인해보면 str이기 때문에 0으로 바뀌지 않는다. 해당 칼럼은 데이터 타입을 int로 바꿀것이 아니기 때문에 괜찮다.

```python
data_fill = data.fillna(0)
data_fill.head()
```

* int 컬럼과 str컬럼을 나눠서 파악하고 astype 으로 컬럼들의 type을 맞춰주도록 하자.

```python
integer_list = list(data_fill.columns)
len(integer_list)
```

```python
string_list = ['1timeline.lane','2timeline.lane','3timeline.lane','4timeline.lane','5timeline.lane',
               '6timeline.lane','7timeline.lane','8timeline.lane','9timeline.lane','10timeline.lane']
```

```python
for i in string_list:
    integer_list.remove(i)
len(integer_list)
```

```python
for i in integer_list:
    data_fill[i] = data_fill[i].astype('float')
for j in string_list:
    data_fill[j] = data_fill[j].astype('str')
```

```python
data_fill.dtypes
```

* boolean과 Win Fail도 0과 1로 나타내며 필요치 않은 칼럼은 drop한 최종적인 데이터 형태를 얻어냈다.

```python
'''
저장
'''
data_fill.to_pickle('data/saving_data/data_info_float.pkl')
```
