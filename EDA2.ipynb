```python
match_fin = pd.DataFrame()
for i in tnrange(len(match_info_loading), desc='1st Loop'):    
    
    api_match_url='https://kr.api.riotgames.com/lol/match/v4/matches/' + str(match_info_loading['gameId'].iloc[i]) + '?api_key=' + api_key
    r_match_info = requests.get(api_match_url)

    if r_match_info.status_code == 200: # response가 정상이면 바로 맨 밑으로 이동하여 정상적으로 코드 실행
        pass

    elif r_match_info.status_code == 429:
        while True: # 429error가 끝날 때까지 무한 루프
            if r_match_info.status_code == 429:
                time.sleep(3)
                r_match_info = requests.get(api_match_url)

            elif r_match_info.status_code == 200: #다시 response 200이면 loop escape
                break

    elif r_match_info.status_code == 503: # 잠시 서비스를 이용하지 못하는 에러
        while True:
            if r_match_info.status_code == 503 or r_match_info.status_code == 429:
                time.sleep(3)
                r_match_info = requests.get(api_match_url)

            elif r_match_info.status_code == 200: # 똑같이 response가 정상이면 loop escape
                break
    elif r_match_info.status_code == 403: # api갱신이 필요
        print('you need api renewal')
        break

    # 위의 예외처리 코드를 거쳐서 내려왔을 때 해당 코드가 실행될 수 있도록 작성
    mat = pd.DataFrame(list(r_match_info.json().values()), index=list(r_match_info.json().keys())).T
    match_fin = pd.concat([match_fin,mat])
```
